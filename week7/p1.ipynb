```python
from scipy import stats
print("SKEWNESS ANALYSIS")
numerical_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
for col in numerical_cols:
    skewness = stats.skew(df[col].dropna())
    print(f"\n{col}:")
    print(f"  Skewness: {skewness:.4f}")
    
    # Interpret skewness
    if skewness > 1:
        print(f"  Interpretation: Highly right-skewed (positively skewed)")
    elif skewness > 0:
        print(f"  Interpretation: Moder
        ately right-skewed")
    elif skewness < -1:
        print(f"  Interpretation: Highly left-skewed (negatively skewed)")
    elif skewness < 0:
        print(f"  Interpretation: Moderately left-skewed")
    else:
        print(f"  Interpretation: Symmetrical (approximately normal)")
```

```python
# Calculate kurtosis for numerical columns
print("\n" + "="*50)
print("KURTOSIS ANALYSIS")
print("="*50)

for col in numerical_cols:
    kurtosis = stats.kurtosis(df[col].dropna())
    print(f"\n{col}:")
    print(f"  Kurtosis: {kurtosis:.4f}")
    
    # Interpret kurtosis (excess kurtosis)
    if kurtosis > 1:
        print(f"  Interpretation: Leptokurtic (heavy tails, sharp peak)")
    elif kurtosis > 0:
        print(f"  Interpretation: Slightly leptokurtic")
    elif kurtosis < -1:
        print(f"  Interpretation: Platykurtic (light tails, flat peak)")
    elif kurtosis < 0:
        print(f"  Interpretation: Slightly platykurtic")
    else:
        print(f"  Interpretation: Mesokurtic (approximately normal)")
```
