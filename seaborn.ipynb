## seaborn - Statistical data visualization

### 1.What is it??
seaborn is a python library that builts on ths top of the matplotlib that makes the statistical data more easy and beautiful

### Why it is used?
For craeting plots like violin,box,paired plots

Automatically handles data from pandas data frames

### Where it is used??
Data Analysis and reporting

Exploring relationships between variables

Machine Learning for visualizing data sets and results





### 2.Import

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load Titanic dataset
df = sns.load_dataset("titanic")
df.head()
 
```

```python
print(df.shape)
```

```python
## 4. Histograms & Countplots
# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
 
```

```python
### 5. Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
 
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
 
```

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
 
```

```python
### 8. Heatmap
# Correlation heatmap (numeric columns only)
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
 
 
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
 
 
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
 
 
```

```python
"Show trends or changes in mean values across categories."
sns.pointplot(x='class', y='fare', data=df, hue='sex', markers=['o', 'x'])
plt.title("Point Plot of Fare by Class and Sex")
plt.show()
 
 
```

```python
sns.kdeplot(df['fare'].dropna(), fill=True, color='teal')
plt.title("KDE Plot of Fare Distribution")
plt.show()
 
 
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='class', y='survived', hue='sex')
plt.title('Passenger Survival by Class and Gender')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='age', bins=20, kde=True)
plt.title('Distribution of Passenger Ages')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='class', y='fare')
plt.title('Fare Distribution by Class')
plt.show()
```

```python

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='age', y='fare', hue='class', palette=['#DC143C', '#3CB371', '#6495ED'])
plt.title('Relationship Between Age and Fare')
plt.show()

```

```python
plt.figure(figsize=(10,8))
corr_matrix = df[['survived', 'age', 'fare']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
